```python
import sqlite3
```

```python
conn = sqlite3.connect('library.db')
cursor = conn.cursor()
```

```python
cursor.execute('''
create table if not exists books(
bookID integer primary key,
title text not null,
author text,
genre text,
published year integer)
''')
```

```python
cursor.execute('delete from books')
conn.commit()
```

```python
cursor.execute('''insert into
books values(1,'Harry potter and the philosophers stone', 'J K Rowling','fiction',2001)''')
```

```python
conn.commit()
```

```python
books = [
(2,'Indulekha','O. Chandu Menon','Early Malayalam fiction',1889),
(3,'Kayar','Thakazhi Sivasankara Pillai','Historical fiction',1978),
(4,'The Alchemist','Paulo Coelho','fiction',1988),
(5,'Balyakalasakhi','Vaikom Muhammad Basheer','romantic tragedy',1944)]
```

```python
cursor.executemany('insert into books values(?,?,?,?,?)',books)
```

```python
conn.commit()
```

```python
cursor.execute('select * from books')
rows = cursor.fetchall()
```

```python
for row in rows:
  print(row)
```
