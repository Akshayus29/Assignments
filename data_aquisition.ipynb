### **Reading data from various sources.**

```python
import numpy as np
import pandas as pd
```

```python
#1st excel file
employees = pd.read_excel('/content/employee.xlsx',header=1,sheet_name = 'Sheet1' )
employees
```

```python
xls = pd.ExcelFile('/content/manager_staff.xlsx')
xls.sheet_names
```

```python
#2nd excel file sheet1=manager

manager = pd.read_excel('/content/manager_staff.xlsx',header=1,sheet_name = 'manager' )
manager
```

```python
#2nd excel file sheet2=staff

staff = pd.read_excel('/content/manager_staff.xlsx',header=1,sheet_name = 'staffs' )
staff
```

```python
combined = pd.concat([manager,staff],ignore_index=True)
combined
```

```python
#to create combined excel file
combined.to_excel('combined.xlsx',index=False)
```

```python
combined.info()
```

```python
combined.groupby('department')['salary'].mean()
```
