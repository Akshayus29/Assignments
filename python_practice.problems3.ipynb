Write a Python program which accepts the radius of a circle from the user
and compute the area.

```python
import math
radius = float(input("Enter the radius of the circle: "))
area_of_circle = math.pi*radius**2
print(area_of_circle)
```

Given a range of first 10 numbers, write a Python program to iterate from
start number to the end number and print the sum of the current number
and previous number.  


```python
previous = 0

for current in range(1, 11):
    total = current + previous
    print("Current Number:", current, "Previous Number:", previous, "Sum:", total)
    previous = current
```

Write a Python program to print only those numbers which are divisible
of 5.

```python
numbers = [10, 20, 33, 46, 55]

for n in numbers:
    if n % 5 == 0:
        print(n)
```

Write a Python program to check whether a number is prime or not.

```python
num = int(input("Enter a number: "))

if num <= 1:
    print(num, "is not a prime number")
else:
    for i in range(2, num):
        if num % i == 0:
            print(num, "is not a prime number")
            break
    else:
        print(num, "is a prime number")
```

 Write a Python program to reverse a list using for loop.  


```python
numbers = [1, 2, 3, 4, 5]

reversed_list = []

for i in range(len(numbers)-1, -1, -1):
    reversed_list.append(numbers[i])

print("Original List:", numbers)
print("Reversed List:", reversed_list)
```

Write a Python program to construct the following pattern, using a nested for loop.
*
**
***
****
*****
****
***
**
*


```python
# Increasing part
for i in range(1, 6):
    for j in range(i):
        print("*", end="")
    print()

# Decreasing part
for i in range(4, 0, -1):
    for j in range(i):
        print("*", end="")
    print()
```
