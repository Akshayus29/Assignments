# Q1. Write a function that return the range (maximum and minimun) of a given list of numbers

```python
def find_range(numbers):
  maximum = max(numbers)
  minimum =min(numbers)
  return maximum,minimum

numbers = [7,5,13,3,9,1,11]
result = find_range(numbers)
print("maximum:" , result[0])
print("minimum:" , result[1])
```

```python
def find_range(list_numbers):
#assuming both 1st number is the maximum and minimum
  minimum = list_numbers[0]
  maximum = list_numbers[0]

#looping thropugh the entire list to check for
#maximum and minimum
  for number in list_numbers:
    if number >= maximum:
      maximum = number
    if number <= minimum:
      minimum = number

  return maximum,minimum

list_numbers = [7,5,13,3,9,1,11]
range = find_range(list_numbers)
print(range)

```

## Q2. Write function that returns mean value of a list of numbers

```python
def mean(numbers):
  total = sum(numbers)
  count = len(numbers)
  mean = total/count
  return mean

numbers = [7,5,13,3,9,1,11]
result = mean(numbers)
result
```

##Q8. Write a function to return standard deviation of a list of numbers

```python
def standard_deviation(numbers):
  n = len(numbers)
  mean = sum(numbers) / n
  variance = sum((x - mean)**2 for x in numbers) / n
  return (variance)*0.5

numbers = [7,5,13,3,9,1,11]
print(standard_deviation)
```
