### **Explore Numpy**

```python
# python list
list1 = [1,2,3,4]
list2 = [5,6,7,8]

print(type(list1))
print(type(list2))
```

```python
# arrays list
array1 = np.array([2,4,6,8])
array2 = np.array([1,3,5,7])

print(type(array1))
print(type(array2))
```

```python
# addding 2 lists
sum_list = list1 + list2
print(sum_list)
```

```python
#element wise addition of 2 lists
sum_list1 = []
for m,n in zip(list1,list2):
  sum_list1.append(m+n)
print(sum_list1)
```

```python
# adding 2 arrays
sum_array = array1 + array2
print(sum_array)
```

```python
# shape of the array is no.of rows,no:of columns
array1.shape
```

```python
#example
item_prices = [100,200,250]
discount = 15
discounted_price = []
#calculating 15% discount for each item
for price in item_prices:
  discount_amount = price * discount / 100
  price_after_discount = price - discount_amount
  discounted_price.append(price_after_discount)

print("discounted prices are :",discounted_price)

#using arrays
item__prices_array =  np.array(item_prices)
print(type(item__prices_array))
#calculating discount amount for each element in the array
discount_amount_array = item__prices_array * discount / 100
discounted_price_array = item__prices_array - discount_amount_array
print("discounted prices are (using arrays) :", discounted_price)

```

```python
dsa_array = np.array ([[22,21,23],[179,160,180]])
csa_array = np.array ([[21,20,19],[157,171,160]])
fsd_array = np.array ([[20,21,30],[167,159,180]])

ict_array = np.array([dsa_array,csa_array,fsd_array])

print("DSA")
print("shape :",dsa_array.shape)
print("size :",dsa_array.size)
print("dimention :",dsa_array.ndim)
print("data type :",dsa_array.dtype)
print("******************************")
print("CSA")
print("shape :",csa_array.shape)
print("size :",csa_array.size)
print("dimention :",csa_array.ndim)
print("data type :",csa_array.dtype)
print("******************************")
print("FSA")
print("shape :",fsd_array.shape)
print("size :",fsd_array.size)
print("dimention :",fsd_array.ndim)
print("data type :",fsd_array.dtype)
print("******************************")
print("ICT")
print("shape :",ict_array.shape)
print("size :",ict_array.size)
print("dimention :",ict_array.ndim)
print("data type :",ict_array.dtype)
```
