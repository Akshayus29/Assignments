```python
import pandas as pd
import numpy as np
```

```python
data = {'First Score':[100,90, np.nan, 95, np.nan],
        'Second Score':[30,45,56,np.nan,np.nan],
        'Third score':[np.nan,40,80,98,np.nan],
        'Fourth Score':[np.nan,np.nan,np.nan,65,np.nan],
        'Fifth Score':[12,15,18,43,54]}

df = pd.DataFrame(data)
df
```

```python
df.isna().sum()
```

```python
df.fillna(1)

#.fillna() allows you to replace these gaps with specific values so your data remains usable for analysis or machine learning.
```

```python
df.ffill()

#.ffill (short for Forward Fill) is a specific method used to propagate the last valid observation forward to next valid.
```

```python
df.bfill()

#.bfill() stands for Backward Fill.
```

```python
replace_vals = {'First Score':10.1,'Second Score':20.1,'Third Score':30.1}
df.fillna(value=replace_vals)
```

```python
df.replace(to_replace=np.nan,value=replace_vals)
```

```python
df.fillna(df.mean())
```

```python
df2 = df.copy()
df2
```

```python
df2['First Score'] = df2['First Score'].fillna(df2['First Score'].mean())
df2
```

```python
df2['Second Score'] = df2['Second Score'].fillna(df2['Second Score'].median())
df2
```

```python
df2.loc[(df2['Second Score'].isna()) & (df['Fifth Score']>20),'Second Score']=999
df2
```

```python
df.dropna()

#.dropna() is the method used to remove missing data entirely.
```

```python
df.dropna(axis=1)
```

```python
cols = ['beer_servings','spirit_servings','wine_servings','total_litres_of_pure_alcohol']

for col in cols:
  beer_serv[col] = beer_serv[col].fillna(beer_serv[col].median())
beer_serv
```

```python
beer_serv.plot(kind='box')
```

```python
#spirit servings

q1 = np.percentile(beer_serv['spirit_servings'],25)
q3 = np.percentile(beer_serv['spirit_servings'],75)
IQR = q3-q1
print(q1,q3,IQR)
```

```python
up_val = q3+1.5*IQR
low_val = q1-1.5*IQR
print(up_val,low_val)
```

```python
for x in beer_serv['spirit_servings']:
  if x >up_val or x < low_val:
    print(x)
```

```python
beer_serv['spirit_servings'] = beer_serv['spirit_servings'].clip(lower=low_val,upper=up_val)


#this is to remove outliers
```

```python
for x in beer_serv['spirit_servings']:
  if x >up_val or x < low_val:
    print(x)

#if we run this there is no output because there is no outliers
```
