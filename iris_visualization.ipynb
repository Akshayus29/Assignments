```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
iris = sns.load_dataset('iris')
iris
```

```python
sns.boxplot(data=iris)

#This function is used to create a Box Plot (also known as a Box-and-Whisker Plot).
```

```python
sns.histplot(iris)
```

```python
sns.kdeplot(iris)

#A KDE plot essentially draws a smooth, continuous curve that represents the distribution of the data, acting as a smoothed-out version of a histogram.
```

```python
sns.pairplot(iris)

 #pairplot is a function in seaborn library, used for see all together
```

```python
sns.pairplot(iris,hue='species')

 # sns.pairplot creates a grid of plots where every numerical variable in the dataset is plotted against every other numerical variable.'hue=' is a parameter used for seeing separately
```

```python
sns.violinplot(iris)

#Combines a box plot with a KDE plot (density). It shows the full distribution shape like a KDE, but also includes the median and quartiles like a box plot. Great for comparing distributions across categories
```

```python
sns.jointplot(iris)

#Joint Distribution Plot | Creates a figure showing the relationship between two variables, typically with a scatterplot in the center and marginal histograms/KDE plots on the x and y axes.
```

```python
sns.ecdfplot(iris)

#Empirical Cumulative Distribution Function | Plots the proportion of data that is less than or equal to a given value. Useful for comparing two distributions and seeing the cumulative frequency.
```

```python
sns.scatterplot(iris)

#he most basic way to visualize the relationship between two or more numerical variables. You can easily add a third variable using the 'hue' or 'size' parameters.
```

```python
sns.lineplot(iris)

#Best used for visualizing how one variable changes over another, typically time or an ordered category. Excellent for time-series data.
```

```python
sns.regplot(data=iris,x='sepal_length',y='sepal_width') #Regression Plot

#Draws a scatter plot and fits a linear regression model (straight line) through the data, often with a shaded confidence interval around the line. Great for checking the linearity of a relationship.
```

```python
sns.barplot(iris)

#Displays the mean (or other aggregate function) of a numerical variable for each category, usually with error bars (confidence intervals). Important: Unlike a histogram, it shows an aggregate measure, not counts.
```

```python
sns.pointplot(iris)

# Similar to a bar plot, but plots the estimate (e.g., mean) as a point and connects the points across categories with lines. Useful for showing interaction effects.
```

```python
sns.stripplot(iris)

#Plots the individual data points for a numerical variable within each category. Useful for small datasets or when combined with a box plot to show the underlying data.
```

```python
sns.swarmplot(iris)

# Similar to a strip plot, but adjusts the points so they don't overlap, creating a "swarm" effect. A great alternative to a strip plot for showing individual observations.
```

```python
sns.countplot(iris)

#Similar to a histogram, but specifically for categorical variables. It shows the count of observations in each category (i.e., it plots the frequency of categories).
```

```python
iris_new = iris.select_dtypes(include=['number'])
corr = iris_new.corr()
corr
```

```python
sns.heatmap(corr,annot=True)
```

```python
sns.heatmap(corr,annot=True,cmap='Greens')

#cmap is a parameter used for color grading ,color started with caps
```

```python
sns.clustermap(iris_new)

#A specialized heatmap that performs hierarchical clustering on the data to rearrange the rows and columns. This helps in visually grouping observations or features that are similar to each other.
```
